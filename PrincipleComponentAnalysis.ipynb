```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv('SkillCraft1_Dataset.csv')

# Drop irrelevant columns (e.g., GameID, LeagueIndex)
df = df.drop(['GameID', 'LeagueIndex'], axis=1)

# Convert non-numeric values to NaN
df = df.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
df = df.dropna()

# Separate the features from the target variable
X = df.drop('Age', axis=1)
y = df['Age']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


```

```python
# Apply PCA with n_component = 2
n = 2
pca = PCA(n_components=n)
X_pca = pca.fit_transform(X_scaled)

# Create a new data frame with the principal components
new_df = pd.DataFrame(data=X_pca, columns=[f"PC{i+1}" for i in range(n)])

# Display the new data frame
new_df.head(5)
```

```python
# Apply PCA with n_component = 5
n = 5
pca = PCA(n_components=n)
X_pca = pca.fit_transform(X_scaled)

# Create a new data frame with the principal components
new_df = pd.DataFrame(data=X_pca, columns=[f"PC{i+1}" for i in range(n)])

# Display the new data frame
new_df.head(5)
```
